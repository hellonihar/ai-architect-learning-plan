```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline

print('Hugging Face libraries loaded')
```

## Tokenization

```python
tokenizer = AutoTokenizer.from_pretrained('gpt2')
text = "Tokenization converts text into numbers."
tokens = tokenizer(text)

print(f'Text: {text}')
print(f'Tokens: {tokenizer.convert_ids_to_tokens(tokens["input_ids"])}')
print(f'IDs: {tokens["input_ids"]}')
```

## Text Generation (GPT-like)

```python
generator = pipeline('text-generation', model='gpt2')
prompt = "The future of AI architecture is"
result = generator(prompt, max_length=50, num_return_sequences=1)
print(result[0]['generated_text'])
```
